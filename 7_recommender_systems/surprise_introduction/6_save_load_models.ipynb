# Save and load models

Some algorithms may take too much time to train. In this cases, we are interested in storing the algorithm to disk and loading them afterwards. This notebook provides an example on how can be done with surprise

This notebook is based on the code from: https://github.com/NicolasHug/Surprise/blob/master/examples/serialize_algorithm.py


## Imports

```python
!pip install scikit-surprise
```

```python
from __future__ import (absolute_import, division, print_function,
                        unicode_literals)
import os

from surprise import SVD
from surprise import Dataset
from surprise import dump
```

# Train the model

```python
data = Dataset.load_builtin('ml-100k')
trainset = data.build_full_trainset()

algo = SVD()
algo.fit(trainset)

# Compute predictions of the 'original' algorithm.
predictions = algo.test(trainset.build_testset())
```

## Store the model

We are storing it in a file named "dump_file"

```python
file_name = os.path.expanduser('~/dump_file')
dump.dump(file_name, algo=algo)
```

## Load the model from disk

```python
_, loaded_algo = dump.load(file_name)
```

# Check that both model obtain the same results

```python
# We now ensure that the algo is still the same by checking the predictions.
predictions_loaded_algo = loaded_algo.test(trainset.build_testset())
assert predictions == predictions_loaded_algo
print('Predictions are the same')
```
